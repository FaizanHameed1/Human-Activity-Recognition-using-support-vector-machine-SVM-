We can see that the data is imbalance because evary calss have different number 
of entries. WALKING_DOWNSTAIRES have minimum number of entries i.e 1406. So, to make data balance we will take all calassis with equal number of samples.

```python
#Dataframe with class 'WALKING_DOWNSTAIRS'
df_wd=df[df['Activity']=='WALKING_DOWNSTAIRS']
df_wd.shape
```

```python
#Dataframe with class 'WALKING_UPSTAIRS' having number of samples equel to 'WALKING_DOWNSTAIRS'
df_wlu = df[df['Activity']== 'WALKING_UPSTAIRS']# Making a saparate dataframe with class "WALKING_UPSTAIRS" in column "Activity".
df_wu=df_wlu.sample(n=df_wd.shape[0])# selecting number of samples (1406) equal to 'WALKING_UPSTAIRS' class for balancing data.
df_wu.shape
```

```python
#Dataframe with class 'WALKING' having number of samples equel to 'WALKING_DOWNSTAIRS'
df_wl=df[df['Activity']=='WALKING']
df_w=df_wl.sample(n=df_wd.shape[0])
df_w.shape
```

```python
#Dataframe with class 'SITTING' having number of samples equel to 'WALKING_DOWNSTAIRS'
df_asi=df[df['Activity']=='SITTING']
df_si=df_asi.sample(n=df_wd.shape[0])
df_si.shape
```

```python
#Dataframe with class 'STANDING' having number of samples equel to 'WALKING_DOWNSTAIRS'
df_ast=df[df['Activity']=='STANDING']
df_st=df_ast.sample(n=df_wd.shape[0])
df_st.shape
```

```python
#Dataframe with class 'LAYING' having number of samples equel to 'WALKING_DOWNSTAIRS'
df_al=df[df['Activity']=='LAYING']
df_l=df_al.sample(n=df_wd.shape[0])
df_l.shape
```

Now if we concatinate all dataframes and than split train and test datasets, some sample can be more in numbers than others that can increse biasness so manually make train and test datasets.

```python
#Making saparate dataframes for 'WALKING_DOWNSTAIRS'
x_wd = pd.DataFrame(df_wd.drop(['Activity','subject'],axis=1))
y_wd= df_wd.Activity.values.astype(object)
x_train_wd,x_test_wd,y_train_wd,y_test_wd = train_test_split(x_wd,y_wd,test_size=0.2,random_state=0)

#Making saparate dataframes for 'WALKING_UPSTAIRS'
x_wu = pd.DataFrame(df_wu.drop(['Activity','subject'],axis=1))
y_wu= df_wu.Activity.values.astype(object)
x_train_wu,x_test_wu,y_train_wu,y_test_wu = train_test_split(x_wu,y_wu,test_size=0.2,random_state=0)

#Making saparate dataframes for "Walking"
x_w = pd.DataFrame(df_w.drop(['Activity','subject'],axis=1))
y_w= df_w.Activity.values.astype(object)
x_train_w,x_test_w,y_train_w,y_test_w = train_test_split(x_w,y_w,test_size=0.2,random_state=0)

#Making saparate dataframes for 'SITTING'
x_si = pd.DataFrame(df_si.drop(['Activity','subject'],axis=1))
y_si= df_si.Activity.values.astype(object)
x_train_si,x_test_si,y_train_si,y_test_si = train_test_split(x_si,y_si,test_size=0.2,random_state=0)

#Making saparate dataframes for 'STANDING'
x_st = pd.DataFrame(df_st.drop(['Activity','subject'],axis=1))
y_st= df_st.Activity.values.astype(object)
x_train_st,x_test_st,y_train_st,y_test_st = train_test_split(x_st,y_st,test_size=0.2,random_state=0)

#Making saparate dataframes for 'LAYING'
x_l = pd.DataFrame(df_l.drop(['Activity','subject'],axis=1))
y_l= df_l.Activity.values.astype(object)
x_train_l,x_test_l,y_train_l,y_test_l = train_test_split(x_l,y_l,test_size=0.2,random_state=0)
```

```python
#Now concatinating all dataframes
x_train = pd.concat([x_train_wd,x_train_wu,x_train_w,x_train_si,x_train_st,x_train_l])
x_test = pd.concat([x_test_wd,x_test_wu,x_test_w,x_test_si,x_test_st,x_test_l])
x_train.head(3),x_test.head(3)
```

```python
#Concatinating the arrays
y_train = np.concatenate((y_train_wd,y_train_wu,y_train_w,y_train_si,y_train_st,y_train_l))
y_test = np.concatenate((y_test_wd,y_test_wu,y_test_w,y_test_si,y_test_st,y_test_l))

#Converting arrays into dataframe
#y_train = pd.DataFrame(y_train, columns=['Activity'])
#y_test = pd.DataFrame(y_test, columns=['Activity'])
#y_train.head(3),y_test.head(3)
```

```python
#Shuffling the dataframes
x_train=shuffle(x_train)
x_test=shuffle(x_test)
y_train=shuffle(y_train)
y_test=shuffle(y_test)
```

```python
encoder=LabelEncoder()
y_train=encoder.fit_transform(y_train)
y_test=encoder.fit_transform(y_test)

print('y_train= ',y_train)
print('y_test= ',y_test)
```

It is important to scale the data first.

```python
scaler=StandardScaler()#scaling data
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
x_train,x_test
```

```python
print('shape of x_train=  ',x_train.shape,',   shape of y_train=  ',y_train.shape,',    shape of x_test=  ',x_test.shape,',    shape of y_test=   ',y_test.shape)
```

```python
#performing grid search for best parameters selection
grid_param = [
  {'C': [1, 10, 100, 1000], 'gamma': [0.001, 0.0001], 'kernel': ['rbf']},
  {'C': [1, 10, 100, 1000], 'gamma': [0.001, 0.0001], 'kernel': ['linear']},
 ]#different types of kernals in SVM are 'rbf','linear' and 'polynomial' kernals
#Tuning best parameters to fit SVM
model = GridSearchCV(SVC(), grid_param,cv=5)#Perform CV to tune best parameters for SVM model
```

```python
model.fit(x_train, y_train)# Training the SVM model

main_model=model.best_estimator_ #model with tuned parameters

print(main_model)
```

```python
#Saving the model
import pickle

filename = 'finalized_model.sav'
pickle.dump(main_model, open(filename, 'wb'))
```

```python
#Reloading the model

loaded_model = pickle.load(open(filename, 'rb'))
```

```python
#predicting
y_pred = loaded_model.predict(x_test)#predicting
```

```python
#Confusion matrix
conf_matrix=confusion_matrix(y_test,y_pred)
print(conf_matrix)
```
